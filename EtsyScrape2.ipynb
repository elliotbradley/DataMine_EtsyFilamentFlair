```python
#Initializing For Code
from urllib.request import urlopen
import csv
url = "https://www.etsy.com/shop/FilamentFlair"
page = urlopen(url)
html_bytes = page.read()
html = html_bytes.decode("utf-8")

#Preparing for Loop
current_html = html

#GET LOOP NUMBER
Iterations = html.count('<p class="shop2-review-attribution">')

#STORE INFO IN CSV
with open('DataSheet1.csv', 'w', encoding='utf-8') as f:
    writer = csv.writer(f)
    writer.writerow(['Name','Date','Rating','Comments','Product'])
    
    for i in range(Iterations):
        # MOVE HTML TO NAME
        start_index = current_html.find('<p class="shop2-review-attribution">')+ len('<p class="shop2-review-attribution">')
        current_html = current_html[start_index:]

        #COLLECT NAME
        start_index = current_html.find('>')+1
        end_index = current_html.find('</a>')
        Name = current_html[start_index:end_index]
        
        #MOVE HTML TO DATE
        current_html = current_html[current_html.find('on ')+len('on '):]
        
        #COLLECT DATE
        start_index = 0
        end_index = current_html.find('\n')
        Date = current_html[start_index:end_index]
        
        #MOVE HTML TO RATING
        current_html = current_html[current_html.find('span class="stars-svg "'):]
        current_html = current_html[current_html.find('name="rating" value=')+len('name="rating" value='):]
        
        #COLLECT RATING
        start_index = 1
        end_index = current_html.find('" />')
        Rating = current_html[start_index:end_index]
        
        #MOVE HTML TO COMMENTS
        current_html = current_html[current_html.find('<p class="prose break-word m-xs-0">')+len('<p class="prose break-word m-xs-0">'):]
        
        #COLLECT COMMENTS
        start_index = 0
        end_index = current_html.find('</p>')
        Comments = current_html[start_index:end_index]
        
        #MOVE HTML TO PRODUCT BOUGHT
        current_html = current_html[current_html.find('<div class="flag-body hide-xs hide-sm">')+len('<div class="flag-body hide-xs hide-sm">\n                    <p>'):]
        
        #COLLECT PRODUCT BOUGHT
        start_index = 0
        end_index = current_html.find('</p>')
        Product = current_html[start_index:end_index]
        
        #WRITE ROW INTO CSV
        writer.writerow([Name,Date,Rating,Comments,Product])
```
